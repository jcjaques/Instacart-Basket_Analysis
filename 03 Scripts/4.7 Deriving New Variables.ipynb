```python
# Creating Price_label function
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Create new column and use busiest days to represent the two busiest days\slowest days to represent two slowest days.
result2 = []
for value in df_ords_prods_merge["orders_day_of_week"]:
    if value == 0 or value == 1:
        result2.append("Busiest Days")
    elif value == 3 or value == 4:
        result2.append("Slowest Days")
    else:
        result2.append("Regularly Busy")
```

## Step 4

```python
# Find name of column to use to find which hours of the day have most, average, and fewest orders
df_ords_prods_merge.head()
```

```python
# Use Value counts function on column order_time_of_day to get how many orders for each hour
df_ords_prods_merge['order_time_of_day'].value_counts(dropna = False)
```

```python
# Because step 4 wants all 24 hours in the day to fall into 3 catagories, I will take the top 8 times with the most orders and the bottom 8 days with the least amount of orders
```

```python
# Create new column to show which times of the day have Most Orders, Fewest Orders, and Average Orders
result3 = []
for value in df_ords_prods_merge["order_time_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result3.append("Most Orders")
    elif value in [3, 4, 2, 5, 1, 0, 6, 23]:
        result3.append("Fewest Orders")
    else:
        result3.append("Average Orders")
```

```python
result3
```

## Step 5

```python
# Change result3 column name to busiest_period_of_day
df_ords_prods_merge['busiest_period_of_day'] = result3
```

```python
# Checking value counts for busiest_period_of_day column
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
