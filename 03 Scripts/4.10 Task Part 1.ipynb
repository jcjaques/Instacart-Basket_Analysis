### Number 4 - Create an exclusion flag for low-activity customers that have lss than 5 orders and exclude them from data.

```python
# Create activity flag for customers with less than 5 orders
df_cop.loc[df_cop['max_order'] < 5, 'activity_flag'] = 'Low-activity Customer'
df_cop.loc[df_cop['max_order'] >= 5, 'activity_flag'] = 'High-activity Customer'
```

```python
# Check value counts
df_cop['activity_flag'].value_counts(dropna = False)
```

### Number 5 - Create a profiling variable based on income, age, certain goods in 'department_id' and number of dependants.

```python
# Create income group variable
df_cop.loc[df_cop['income'] <= 55000, 'income_group'] = 'low-income'
df_cop.loc[(df_cop['income'] > 55000) & (df_cop['income'] <= 150000), 'income_group'] = 'middle-income'
df_cop.loc[df_cop['income'] > 150000, 'income_group'] = 'high-income'
```

```python
# Check value counts to ensure income groups created.
df_cop['income_group'].value_counts(dropna = False)
```

```python
# Create age group variable using young adult, adult, middle-age, and senior designations.
df_cop.loc[df_cop['age'] <= 25, 'age_group'] = 'young adult'
df_cop.loc[(df_cop['age'] > 25) & (df_cop['age'] < 40), 'age_group'] = 'adult'
df_cop.loc[(df_cop['age'] >= 40) & (df_cop['age'] <= 60), 'age_group'] = 'middle-aged'
df_cop.loc[df_cop['age'] > 60, 'age_group'] = 'senior'
```

```python
# Create value counts to ensure age groups created
df_cop['age_group'].value_counts(dropna = False)
```

```python
# Create dependant variable group
df_cop.loc[df_cop['number_of_dependants'] == 0, 'dependants_flag'] = 'No Dependants'
df_cop.loc[df_cop['number_of_dependants'] >= 1, 'dependants_flag'] = 'Yes Dependants'
```

```python
# Create value counts to ensure dependant groups created
df_cop['dependants_flag'].value_counts(dropna = False)
```
